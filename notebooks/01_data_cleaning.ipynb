## STEP 2 — Import Libraries

## STEP 13 — Drop Missing Rows

```python
df = df.dropna()
```

## STEP 14 — Confirm Missing Values Removed

```python
df.isnull().sum()
```

## STEP 15 — Duplicate Records Count

```python
df.duplicated().sum()
```

## STEP 16 — Remove Duplicates

```python
df = df.drop_duplicates()
```

## STEP 17 — Standardize Column Names

```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)
df.columns
```
